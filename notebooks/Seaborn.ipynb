```python
fig, ax = plt.subplots()
tips = pd.read_csv("https://raw.githubusercontent.com/mwaskom/seaborn-data/master/tips.csv")
ax.violinplot(tips["total_bill"], vert=False)
plt.show()
```

```python
# Load iris data
iris = sns.load_dataset("iris")
# Set context to paper, too have notebook, talk, poster
sns.set_context("paper", font_scale=2, rc={"font.size":10,"axes.labelsize":20})

sns.swarmplot(x="species", y="petal_length", data=iris)
plt.show()
```

```python
from matplotlib.colors import ListedColormap
# Define a variable N
N = 500

# Construct the colormap
current_palette = sns.color_palette("inferno", n_colors=5)
cmap = ListedColormap(sns.color_palette(current_palette).as_hex())

# Initialize the data
data1 = np.random.randn(N)
data2 = np.random.randn(N)
# Assume that there are 5 possible labels
colors = np.random.randint(0,5,N)

# Create a scatter plot
plt.scatter(data1, data2, c=colors, cmap=cmap)

# Add a color bar
plt.colorbar()

# Show the plot
plt.show()
```

```python
sns.set()
# Set axes style to white for first subplot
with sns.axes_style("white"):
    plt.subplot(211)
    sns.swarmplot(x="species", y="petal_length", data=iris)

# Initialize second subplot
plt.subplot(212)

# Plot violinplot
sns.violinplot(x = "total_bill", data=tips)

# Show the plot                   
plt.show()
```

## Figure Size

```python
# Initialize Figure and Axes object
fig, ax = plt.subplots(figsize=(10,4))

# Load in the data
iris = sns.load_dataset("iris")

# Create swarmplot
sns.swarmplot(x="species", y="petal_length", data=iris, ax=ax)

# Show plot
plt.show()
```

Para otras figuras podemos usar los parametros height y aspect que da la proporción entre el algo y el ancho.

```python
# Set up a factorplot
g = sns.catplot(
    "class",
    "survived",
    "sex",
    data=titanic,
    kind="bar",
    height=4,
    aspect=2,
    palette="viridis",
    legend=True
)

# Show plot
plt.show()
```

## Rotate label and limit axes

```python
# Initialize the data
x = 10 * np.arange(1, 10)
y = x * 2
data = pd.DataFrame(data={'x': x, 'y': y})

# Create an lmplot
grid = sns.lmplot('x', 'y', data, height=4, truncate=True, scatter_kws={"s": 100})

# Rotate the labels on x-axis
grid.set_xticklabels(rotation=30)

# Get hold of the `Axes` objects
axes = grid.ax

# Tweak the `Axes` properties
axes.set_ylim(-10,)
axes.set_xlim(0,150)

plt.title('Mi hermoso titulo')

plt.show()
```

```python
# Create the boxplot
ax = sns.boxplot(x="total_bill", data=tips)

# Set the `xlim`
ax.set(xlim=(0, 100))

# Show the plot
plt.show()
```

## Reescalando los ejes

```python
# Create the data
x = 10 ** np.arange(1, 10)
y = x * 2
data = pd.DataFrame(data={'x': x, 'y': y})

# Initialize figure and ax
fig, ax = plt.subplots()

# Set the scale of the x-and y-axes
ax.set(xscale="log", yscale="log")

# Create a regplot  escala logaritmica
sns.regplot("x", "y", data, ax=ax, scatter_kws={"s": 100})

# Show plot
plt.show()
```
