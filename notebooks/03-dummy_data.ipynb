```python
import numpy as np
import pandas as pd

data = pd.DataFrame(
    {
        'uno': np.random.randn(10),
        'dos': 1.5 + 2 * np.random.randn(10),
        'tres': np.random.uniform(5, 32, 10)
    },
    index = range(11, 21)
)
data
```

## Agregación por categorías

```python
gender = ['male', 'female']
income = ['poor', 'middel class', 'rich']

n = 500
gender_data = []
income_data = []

for i in range(n):
    gender_data.append(np.random.choice(gender))
    income_data.append(np.random.choice(income))
    
```

```python
height = 160 + 30*np.random.randn(n) 
weight = 65 + 10*np.random.randn(n)
age = 30 + 12*np.random.randn(n)
```

```python
data = pd.DataFrame({
    'gender': gender_data,
    'economic': income_data,
    'height': height,
    'weight': weight,
    'age': age
})

data.sample(4)
```

### Agrupación

```python
grouped = data.groupby('gender')
grouped.groups
# Muestra el indice de la fila donde esta.
```

```python
for name, groups in grouped:
    print(name)
    print(groups)
```

```python
grouped.get_group('female')
```

### Operaciones

```python
grouped = data.groupby(['gender', 'economic'])
```

```python
grouped.sum()
```

```python
grouped.mean()
```

```python
grouped.size()
```

```python
grouped.aggregate({
    'age': np.mean,
    'height': np.std
})
```

```python
# Tipificar la edad mean/std
grouped.aggregate({
    'age': lambda x: np.mean(x)/np.std(x)
})
```
