```python
t = np.array([i for i in np.arange(1/40, 18/40, 1/40)])
v = np.array([1, 3.2, 5.1, 7.2, 8.8, 11, 13.4, 15.1, 16.8, 19.4, 20.8, 23.3, 25.1, 27.3, 28.8, 31, 33.2])
plt.plot(t, v, 'go')
plt.show()
```

```python
data = pd.DataFrame({
    't': t,
    'v': v,
})
```

```python
import statsmodels.formula.api as smf

lm = smf.ols(formula="v~t", data=data).fit()
```

```python
lm.params
```

```python
vp = -0.913971 + 80.009804*t
```

```python
plt.plot(t, vp, dashes=[2, 4])
plt.title('Linear Model', size='18')
plt.plot(t, v, 'go')
plt.show()
```

### SkLearn

```python
from sklearn.linear_model import LinearRegression
```

```python
t = t[:, np.newaxis]
lm = LinearRegression().fit(t, v)
```

```python
plt.plot(t, v, 'go')
plt.plot(t, lm.predict(t), 'b', dashes=[2, 4])
plt.title('Linear Model', size='18')
plt.show()
```
