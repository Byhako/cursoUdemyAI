```python
import numpy as np
import pandas as pd
from functools import reduce
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style="darkgrid")
```

```python
data = pd.read_csv('./libro.csv', sep=';')
data.head()
```

```python
numeros = pd.DataFrame([data['n1'], data['n2'], data['n3'], data['n4'], data['n5'], data['n6']])
numeros
```

```python
n1 = [i for i in numeros.iloc[0]];
n2 = [i for i in numeros.iloc[1]];
n3 = [i for i in numeros.iloc[2]];
n4 = [i for i in numeros.iloc[3]];
n5 = [i for i in numeros.iloc[4]];
n6 = [i for i in numeros.iloc[5]];

n = n1 + n2 + n3 + n4 + n5 + n6
len(n)
```

```python
results = []
for i in range(1, 46):
    cont = 0
    for j in n:
        if i == j:
            cont += 1
    results.append(cont)            
```

```python
reduce((lambda x, y: x + y), results)
```

```python
ind = range(1, 46)    # the x locations for the groups
width = 0.35       # the width of the bars: can also be len(x) sequence

plt.figure(figsize=(12, 6), dpi=80)
p1 = plt.bar(ind, results, width)


plt.ylabel('Scores')
plt.title('Scores by group and gender')
plt.xticks(ind, ind)
plt.yticks(np.arange(0, max(results), 5))

plt.show()
```

```python
jugar = []
nuevo = [i for i in results]
claves = [{index:i} for index, i in enumerate(results)]
for i in range(6):
    m = max(nuevo)
    index = nuevo.index(m)
    nuevo.pop(index)
    
    for j in range(45):
        if claves[j][j] == m:
            jugar.append(j+1)
            claves[j][j] = 0
            break

jugar    
```
